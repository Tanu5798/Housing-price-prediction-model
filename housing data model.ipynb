```python
# data cleaning
# Here id columns is unnecessary hence we will drop it
```

```python
df.drop(['Id'],
       axis=1,
       inplace=True)
```

```python
df.info()
```

```python
#Replacing SalePrice empty values with their mean values to make the data distribution symmetric.
```

```python
df['SalePrice'] = df['SalePrice'].fillna(df['SalePrice'].mean())
```

```python
#Drop records with null values (as the empty records are very less).
```

```python
new_df = df.dropna()
```

```python
new_df.isnull().sum()
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
s = (new_df.dtypes == 'object')
object_cols = list(s[s].index)
print("categorical varibles:")
print(object_cols)
print('No of categorical features: ',
     len(object_cols))
```

```python
#Then once we have a list of all the features. We can apply OneHotEncoding to the whole list.
```

```python
OHo_encoder = OneHotEncoder(sparse_output=False)
OHo_cols = pd.DataFrame(OHo_encoder.fit_transform(new_df[object_cols]))
OHo_cols.index = new_df.index
OHo_cols.columns = OHo_encoder.get_feature_names_out()
df_final = new_df.drop(object_cols, axis = 1)
df_final = pd.concat([df_final, OHo_cols], axis=1)
```

```python
df_final
```

```python
#X and Y splitting (i.e. Y is the SalePrice column and the rest of the other columns are X)
```

```python
from sklearn.metrics import mean_absolute_error
from sklearn.model_selection import train_test_split
```

```python
X = df_final.drop(['SalePrice'], axis=1)
Y = df_final['SalePrice']
```

```python
X.head()
```

```python
Y
```

```python
# Split the training set into # training and validation set
```

```python
X_train, X_valid, Y_train, Y_valid = train_test_split(X, Y, train_size=0.8, test_size=0.2, random_state=0)
```

```python
#As we have to train the model to determine the continuous values, so we will be using these regression models.
#SVM-Support Vector Machine
#Random Forest Regressor
#Linear Regressor
```

```python
#Support Vector Machine
#SVM can be used for both regression and classification model. It finds the hyperplane in the n-dimensional plane.
```

```python
from sklearn import svm
from sklearn.svm import SVC
from sklearn.metrics import mean_absolute_percentage_error
```

```python
model_SVR = svm.SVR()
model_SVR.fit(X_train,Y_train)
Y_pred = model_SVR.predict(X_valid)


print(mean_absolute_percentage_error(Y_valid, Y_pred))
```

```python
#Random Forest Regression
#Random Forest is an ensemble technique that uses multiple of decision trees and can be used for both regression and classification tasks.
```

```python
from sklearn.ensemble import RandomForestRegressor
model_RFR = RandomForestRegressor(n_estimators=10,random_state =0)
model_RFR.fit(X_train, Y_train)
Y_pred = model_RFR.predict(X_valid)
mape = mean_absolute_percentage_error(Y_valid, Y_pred)
print('Mean_absolute_percentage_error:',mape)
```

```python
#Linear Regression
from sklearn.linear_model import LinearRegression
```

```python
model_LR = LinearRegression()
model_LR.fit(X_train, Y_train)
Y_pred = model_LR.predict(X_valid)
mape = mean_absolute_percentage_error(Y_valid, Y_pred)
print('Mean_absolute_percentage_error: ', mape)
```
